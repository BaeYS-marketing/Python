```python
import xgboost as xgb
xgb.__version__
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_attri = cancer.data
y = cancer.target
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from xgboost import plot_importance
import warnings 
warnings.filterwarnings('ignore')
```

```python
cancer_df = pd.DataFrame(data=X_attri, columns=cancer.feature_names)
cancer_df['target'] = y
cancer_df
```

```python
cancer_df.values #numpy 형태로 변환도 가능
```

```python
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target, test_size=0.2,
                                                 stratify=cancer.target, random_state=42)

X_train_df = pd.DataFrame(data=X_train, columns=cancer.feature_names)
y_train_df = pd.DataFrame(data=y_train)

X_test_df = pd.DataFrame(data=X_test, columns=cancer.feature_names)
y_test_df = pd.DataFrame(data=y_test)
```

```python
# read in data
dtrain = xgb.DMatrix(data=X_train_df, label=y_train_df)
dtest = xgb.DMatrix(data=X_test_df, label=y_test_df)

# specify parameters via map

#어큐러시 높이기 위한 파라미터 : 하이퍼파라미터 조정
param = {'max_depth':2, 'eta':1, 'objective':'binary:logistic' } 

num_round = 2
bst = xgb.train(param, dtrain, num_round)

# make prediction
preds = bst.predict(dtest)
```

```python
#XGBoost 수행코드 : https://xgboost.readthedocs.io/en/latest/get_started.html
#Dmatrix 자료구조 : https://xgboost.readthedocs.io/en/latest/python/python_intro.html
```

```python
# read in data
dtrain = xgb.DMatrix(data=X_train_df, label=y_train_df)
dtest = xgb.DMatrix(data=X_test_df, label=y_test_df)

# specify parameters via map

#어큐러시 높이기 위한 파라미터 : 하이퍼파라미터 조정
params = {'max_depth':3,     #가중치모델 깊이 : 5를 넘지 않는다.
          'eta':0.1,        #세타 : learning-rate , 경사하강법에서 기울기에 붙는 알파값
          'objective':'binary:logistic',   #objective : 모델목적 : 목표변수의 형태에 따라 ~> 이진분류 
          'eval_metric':'logloss',     #오류함수의 평가성능지표 : cost_function ~> 범주형이니까 logloss로 하겠다. 
          'early_stopping':100         #오류함수 성능개선이 없을 때 조기중단. (최소반복횟수)
                                           #-> 100번 반복해서 성능개선 없을 때  
         } 

num_rounds = 400           #부스팅의 반복횟수

wlist = [(dtrain,'train_data'),(dtest,'validataion_data')]  
#cost_function과 더불어서 모델적합
#모델을 만드는 중간중간 validation_data를 통해 모델평가해서 조정한다.
#수능 전에 모의고사 보는 느낌_ 학습방향/정확도 조절
#훈련데이터를 통한 손실함수값 vs 테스트데이터를 통한 손실함수값 

xgb_model = xgb.train(params = params, \
                     dtrain=dtrain , num_boost_round=num_rounds, evals=wlist)
```

```python
wlist
```

```python
# make prediction
pred_probs=xgb_model.predict(dtest)
preds = [ 1 if x>0.5 else 0 for x in pred_probs]

preds
```
